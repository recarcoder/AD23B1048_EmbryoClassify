```python
import warnings
warnings.filterwarnings('ignore')

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import numpy as np
import pandas as pd
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

from torchvision import transforms
from torchvision.models import (
    mobilenet_v2,    MobileNet_V2_Weights,
    vgg16,           VGG16_Weights,
    vgg19,           VGG19_Weights,
    efficientnet_b3, EfficientNet_B3_Weights,
)
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import seaborn as sns

from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True

# pip install coral-pytorch
from coral_pytorch.losses import corn_loss
from coral_pytorch.dataset import corn_label_from_logits

COMPUTE_DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", COMPUTE_DEVICE)

ROOT_PATH = "/kaggle/input/datasets/abhishekbuddiga06/embryo-dataset"
IMAGE_DIR = os.path.join(ROOT_PATH, "embryo_dataset", "embryo_dataset")
LABEL_DIR = os.path.join(ROOT_PATH, "embryo_dataset_annotations", "embryo_dataset_annotations")

PHASE_NAMES = [
    'tPB2', 'tPNa', 'tPNf', 't2', 't3', 't4', 't5', 't6',
    't7', 't8', 't9+', 'tM', 'tSB', 'tB', 'tEB', 'tHB'
]
phase_to_idx = {p: i for i, p in enumerate(PHASE_NAMES)}
NUM_CLASSES  = 15  # label 15 merged into 14

IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]
MINI_BATCH    = 8
NUM_EPOCHS    = 15


class EmbryoImageDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe.reset_index(drop=True)
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        for _ in range(5):
            try:
                record = self.dataframe.iloc[idx]
                img    = Image.open(record["image"]).convert("RGB")
                img    = img.resize((256, 256))
                if self.transform:
                    img = self.transform(img)
                target = torch.tensor(record["label"], dtype=torch.long)
                return img, target
            except Exception:
                idx = np.random.randint(0, len(self.dataframe))
        return torch.zeros(3, 224, 224), torch.tensor(0)


class CornOrdinalLoss(nn.Module):
    # CORN loss + weighted CE blend to handle class imbalance
    def __init__(self, alpha, num_cls, ce_weight=0.3):
        super().__init__()
        self.num_cls   = num_cls
        self.ce_weight = ce_weight
        self.ce        = nn.CrossEntropyLoss(weight=alpha, label_smoothing=0.05)

    def forward(self, logits_corn, logits_ce, targets):
        return corn_loss(logits_corn, targets, self.num_cls) + \
               self.ce_weight * self.ce(logits_ce, targets)


class DualHeadModel(nn.Module):
    # Shared backbone → CORN head (ordinal) + CE head (imbalance)
    def __init__(self, backbone, feat_dim, num_cls, dropout=0.3):
        super().__init__()
        self.backbone  = backbone
        self.corn_head = nn.Sequential(nn.Dropout(dropout), nn.Linear(feat_dim, num_cls - 1))
        self.ce_head   = nn.Sequential(nn.Dropout(dropout), nn.Linear(feat_dim, num_cls))

    def forward(self, x):
        feats = self.backbone(x)
        return self.corn_head(feats), self.ce_head(feats)


def build_model(arch):
    if arch == "mobilenet":
        base = mobilenet_v2(weights=MobileNet_V2_Weights.DEFAULT)
        feat_dim = base.last_channel
        base.classifier = nn.Identity()
    elif arch == "vgg16":
        base = vgg16(weights=VGG16_Weights.DEFAULT)
        feat_dim = 4096
        base.classifier = nn.Sequential(*list(base.classifier.children())[:-1])
    elif arch == "vgg19":
        base = vgg19(weights=VGG19_Weights.DEFAULT)
        feat_dim = 4096
        base.classifier = nn.Sequential(*list(base.classifier.children())[:-1])
    elif arch == "efficientnet":
        base = efficientnet_b3(weights=EfficientNet_B3_Weights.DEFAULT)
        feat_dim = base.classifier[1].in_features
        base.classifier = nn.Identity()
    else:
        raise ValueError(f"Unknown arch: {arch}")
    return DualHeadModel(base, feat_dim, NUM_CLASSES).to(COMPUTE_DEVICE)


def build_dataframe(step=6, max_embryos=None, max_per_embryo=None):
    records = []
    embryo_count = 0
    for fname in tqdm(os.listdir(LABEL_DIR)):
        if max_embryos and embryo_count >= max_embryos:
            break
        if not fname.endswith(".csv"):
            continue
        emb_id     = fname.replace("_phases.csv", "")
        ann_path   = os.path.join(LABEL_DIR, fname)
        img_folder = os.path.join(IMAGE_DIR, emb_id)
        if not os.path.exists(img_folder):
            continue
        img_list = sorted(os.listdir(img_folder))
        if not img_list:
            continue
        phase_df          = pd.read_csv(ann_path, header=None)
        phase_df.columns  = ["phase", "start", "end"]
        emb_samples = 0
        for _, row in phase_df.iterrows():
            if max_per_embryo and emb_samples >= max_per_embryo:
                break
            cls_idx = phase_to_idx[row["phase"]]
            for frame_no in range(int(row["start"]), int(row["end"]), step):
                if max_per_embryo and emb_samples >= max_per_embryo:
                    break
                if frame_no < len(img_list):
                    records.append([os.path.join(img_folder, img_list[frame_no]), cls_idx, emb_id])
                    emb_samples += 1
        embryo_count += 1
    return pd.DataFrame(records, columns=["image", "label", "embryo"])


def run_train_epoch(net, loader, optimizer, criterion):
    net.train()
    running_loss = 0
    all_preds, all_targets = [], []
    for imgs, labels in tqdm(loader, desc="Training", leave=False):
        imgs, labels = imgs.to(COMPUTE_DEVICE), labels.to(COMPUTE_DEVICE)
        optimizer.zero_grad()
        logits_corn, logits_ce = net(imgs)
        loss = criterion(logits_corn, logits_ce, labels)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(net.parameters(), 1.0)
        optimizer.step()
        running_loss += loss.item()
        batch_preds = corn_label_from_logits(logits_corn)
        all_preds.extend(batch_preds.cpu().numpy())
        all_targets.extend(labels.cpu().numpy())
    return running_loss / len(loader), accuracy_score(all_targets, all_preds)


def run_evaluation(net, loader):
    net.eval()
    all_preds, all_targets = [], []
    with torch.no_grad():
        for imgs, labels in loader:
            logits_corn, _ = net(imgs.to(COMPUTE_DEVICE))
            all_preds.extend(corn_label_from_logits(logits_corn).cpu().numpy())
            all_targets.extend(labels.numpy())
    acc = accuracy_score(all_targets, all_preds)
    f1  = f1_score(all_targets, all_preds, average="weighted", zero_division=0)
    return acc, f1, all_preds, all_targets


def quadratic_weighted_kappa(y_true, y_pred, num_cls):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    W = np.array([[(i - j) ** 2 / (num_cls - 1) ** 2
                   for j in range(num_cls)] for i in range(num_cls)], dtype=np.float64)
    hist_t = np.bincount(y_true, minlength=num_cls).astype(np.float64)
    hist_p = np.bincount(y_pred, minlength=num_cls).astype(np.float64)
    E      = np.outer(hist_t, hist_p) / hist_t.sum()
    O      = np.zeros((num_cls, num_cls), dtype=np.float64)
    for t, p in zip(y_true, y_pred):
        O[t, p] += 1
    denom = (W * E).sum()
    return 1.0 - (W * O).sum() / denom if denom != 0 else 1.0


# ---------- build dataset ----------
full_df = build_dataframe(max_embryos=10, max_per_embryo=200)
print(f"Total samples: {len(full_df)} | Unique embryos: {full_df['embryo'].nunique()}")

unique_embryos = full_df["embryo"].unique()
if len(unique_embryos) < 2:
    train_df, val_df = train_test_split(full_df, test_size=0.2, random_state=42, stratify=full_df['label'])
else:
    train_ids, val_ids = train_test_split(unique_embryos, test_size=0.2, random_state=42)
    train_df = full_df[full_df.embryo.isin(train_ids)].copy()
    val_df   = full_df[full_df.embryo.isin(val_ids)].copy()

# merge rare last phase into second-last
train_df.loc[train_df["label"] == 15, "label"] = 14
val_df.loc[val_df["label"]   == 15, "label"]   = 14
print(f"Train: {len(train_df)} | Val: {len(val_df)}")

raw_weights = compute_class_weight('balanced', classes=np.arange(NUM_CLASSES), y=train_df["label"].values) \
              if len(train_df) > 0 else np.ones(NUM_CLASSES)
cls_weights = torch.tensor(raw_weights, dtype=torch.float32).to(COMPUTE_DEVICE)

# CenterCrop before ToTensor (PIL-level); ImageNet stats match pretrained backbones
augment_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.CenterCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.05, hue=0.02),
    transforms.RandomGrayscale(p=0.1),
    transforms.GaussianBlur(3, sigma=(0.1, 1.5)),
    transforms.ToTensor(),
    transforms.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD),
    transforms.RandomErasing(p=0.2),   # must be after ToTensor
])

eval_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD),
])

train_loader = DataLoader(EmbryoImageDataset(train_df, augment_transform), batch_size=MINI_BATCH, shuffle=True,  num_workers=2)
val_loader   = DataLoader(EmbryoImageDataset(val_df,   eval_transform),    batch_size=MINI_BATCH, shuffle=False, num_workers=2)

# ---------- training loop ----------
perf_results = {}

for arch_name in ["mobilenet", "vgg16", "vgg19"]:
    print(f"\n🚀 Training {arch_name.upper()}")
    net       = build_model(arch_name)
    opt       = torch.optim.Adam(net.parameters(), lr=1e-4, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=NUM_EPOCHS)
    criterion = CornOrdinalLoss(cls_weights, NUM_CLASSES).to(COMPUTE_DEVICE)

    best_f1, best_weights = 0.0, None

    for ep in range(NUM_EPOCHS):
        ep_loss, ep_acc       = run_train_epoch(net, train_loader, opt, criterion)
        v_acc, v_f1, _, _     = run_evaluation(net, val_loader)
        scheduler.step()
        print(f"Epoch {ep+1}/{NUM_EPOCHS} | Loss: {ep_loss:.4f} | Train Acc: {ep_acc:.4f} | Val Acc: {v_acc:.4f} | Val F1: {v_f1:.4f}")
        if v_f1 > best_f1:
            best_f1      = v_f1
            best_weights = {k: v.clone() for k, v in net.state_dict().items()}

    net.load_state_dict(best_weights)
    final_acc, final_f1, final_preds, final_targets = run_evaluation(net, val_loader)
    qwk = quadratic_weighted_kappa(final_targets, final_preds, NUM_CLASSES)
    perf_results[arch_name] = {"acc": final_acc, "f1": final_f1, "qwk": qwk, "net": net}
    print(f"✅ {arch_name.upper()} → Acc: {final_acc:.4f} | F1: {final_f1:.4f} | QWK: {qwk:.4f}")

# ---------- results table ----------
print(f"\n{'Model':<15} {'Acc':>8} {'F1':>8} {'QWK':>8}")
print("-" * 42)
for arch_name, r in perf_results.items():
    print(f"{arch_name.upper():<15} {r['acc']:>8.4f} {r['f1']:>8.4f} {r['qwk']:>8.4f}")

# ---------- bar plots ----------
arch_names = list(perf_results.keys())
fig, axes  = plt.subplots(1, 3, figsize=(14, 4))
for ax, key, label, color in zip(
    axes,
    ["acc", "f1", "qwk"],
    ["Validation Accuracy", "Validation F1 (weighted)", "Quadratic Weighted Kappa"],
    ["steelblue", "steelblue", "seagreen"]
):
    ax.bar(arch_names, [perf_results[m][key] for m in arch_names], color=color)
    ax.set_title(label)
    ax.set_ylim(0, 1)
plt.tight_layout()
plt.show()

# ---------- confusion matrix (best model by QWK) ----------
best_arch = max(perf_results, key=lambda m: perf_results[m]["qwk"])
_, _, final_preds, final_targets = run_evaluation(perf_results[best_arch]["net"], val_loader)

if final_preds:
    cm = confusion_matrix(final_targets, final_preds)
    plt.figure(figsize=(12, 9))
    sns.heatmap(cm, cmap="Blues", annot=True, fmt='d',
                xticklabels=PHASE_NAMES[:NUM_CLASSES],
                yticklabels=PHASE_NAMES[:NUM_CLASSES])
    plt.title(f"Confusion matrix — {best_arch.upper()} (best QWK)")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
